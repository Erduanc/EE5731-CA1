# Part 2: SIFT Features and Descriptors
Extract SIFT features by cv2 provided functions:

```python
import numpy as np
import cv2
from PIL import Image
from pylab import *
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
image = cv2.imread('im01.jpg')
siftCreater = cv2.SIFT_create()

keypoints, descriptors = siftCreater.detectAndCompute(image, None)

newImage = cv2.drawKeypoints(image=image, keypoints=keypoints, outImage=image, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(cv2.cvtColor(newImage,cv2.COLOR_BGR2RGB), interpolation='none')
plt.show()
```
